# Now Getting Exploratory Data Analysis

```python
df.head()
```

```python
df['IncreasingStress'].value_counts()
```

The high number of ‘maybe’ responses suggests uncertainty and lack of awareness, meaning many individuals may not fully recognize or acknowledge their stress levels.

Does Work Interest Creates Stress ?

```python
pd.crosstab(df['WorkInterest'],df['IncreasingStress'],normalize= 'index')
```

Individuals with low work interest show the highest proportion of uncertainty in stress levels, indicating disengagement may contribute to mental instability 


Interestingly, individuals with high work interest also report significant stress levels, suggesting that stress is not solely dependent on interest but may be influenced by workload or external pressures

Is Staying Indoor creates stress ?


```python
pd.crosstab(df['DaysIndoors'],df['IncreasingStress'],normalize='index')
```

Individuals spending more than two months indoors show the highest levels of stress, indicating a strong link between prolonged isolation and declining mental well-being.

People who go out daily report the lowest stress levels, suggesting that regular outdoor activity may help reduce stress.



Moderate indoor duration (31–60 days) shows relatively balanced stress levels, suggesting that extreme isolation, rather than indoor time itself, is a key contributor to stress.

Is Coping cause stress?

```python
pd.crosstab(df['CopingStruggles'], df['IncreasingStress'], normalize='index')
```

Coping struggles do not show a strong direct correlation with increasing stress, as stress levels remain relatively similar across both groups.

is family history affect treatment ?

```python
pd.crosstab(df['FamilyHistory'], df['Treatment'], normalize='index')
```

Individuals with a family history of mental health issues are significantly more likely to seek treatment (72.5%) compared to those without such history (34.7%), indicating that prior exposure increases awareness and willingness to take action.

Do care options affect whether people take treatment?

```python
pd.crosstab(df['CareOptions'], df['Treatment'], normalize='index')
```

Access to or awareness of care options significantly increases the likelihood of seeking treatment, with nearly 70% of individuals opting for treatment when care options are available

# Recommendations

1. Increase Awareness of Mental Health Resources
Organizations should actively promote available mental health support systems, as lack of awareness is a major barrier to treatment.

2. Encourage Work-Life Balance and Outdoor Activities
Reducing prolonged indoor stays through flexible work policies and encouraging social or outdoor engagement can help lower stress levels.

3. Provide Accessible Mental Health Support Systems
Companies should implement counseling services, employee assistance programs, and easy access to mental health professionals.

4. Target High-Risk Groups for Early Intervention
Individuals showing uncertainty (“maybe” responses) should be prioritized for awareness and support programs, as they represent a vulnerable group.

# Conclusion

This analysis reveals that mental health challenges are widespread and influenced more significantly by lifestyle factors such as prolonged isolation than by work-related engagement alone. Among all factors, access to and awareness of mental health care options emerged as the most critical driver of treatment-seeking behavior. The findings highlight the importance of not only providing mental health support but also ensuring that individuals are well-informed and encouraged to utilize these resources. Addressing these areas can lead to improved well-being and more proactive mental health management
